# Car Object detection using image ai
This image detection using Yolo's Algorithm

```python
!pip install imageai
from imageai.Detection import ObjectDetection
```

```python
detector = ObjectDetection()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model_path = "drive/MyDrive/ObjectDetection/models/yolo-tiny.h5"
input_path = "drive/MyDrive/ObjectDetection/input/test45.jpg"
output_path = "drive/MyDrive/ObjectDetection/output/newimage.jpg"
```

```python
detector.setModelTypeAsTinyYOLOv3()
detector.setModelPath(model_path)
detector.loadModel()
detection = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path)
```

```python
for eachItem in detection:
    print(eachItem["name"] , " : ", eachItem["percentage_probability"])
```
